## 함수화

```python
#브라우저 띄우는 함수 

def _browser(headless=False):
    chrome_options = Options()
    if headless is True:
        chrome_options.add_argument("--headless")
    browser = webdriver.Chrome('chromedriver', options=chrome_options)
    browser.implicitly_wait(3)
    browser.set_window_size(1920, 1280)
    return browser
    
def _get_page(url):
    browser.get(url)
    
def login_evertime(browser, _id,_pw):
    ele = browser.find_element_by_css_selector('input[name="userid"]')
    p_ele = browser.find_element_by_css_selector('input[name="password"]')
    ele.send_keys(_id)
    p_ele.send_keys(_pw)
    submit_ele = browser.find_element_by_css_selector('#container > form > p.submit > input')
    submit_ele.click()
    
def _click_hot_post(browser, first_or_second):
    """
    실시간 인기글은 2가지인데 둘중하나 선택해서 이동 후 html return
    """
    if first_or_second == 'first':
        hot_ele = browser.find_element_by_css_selector('#container > div.rightside > div:nth-child(2) > div > a:nth-child(2)')
    elif first_or_second =='second':
        hot_ele = browser.find_element_by_css_selector('#container > div.rightside > div:nth-child(2) > div > a:nth-child(3)')
    hot_ele.click()
    time.sleep(2)
    return browser.page_source

def data_parser(page_source):
    soup = BeautifulSoup(page_source, "html.parser")
    post = soup.select_one('#container > .wrap.articles > article > a.article')
    title = post.h2.text
    contents=""
    for p in post.select('p'):
        contents += p.text
    vote = post.select_one('#container > div.wrap.articles > article > a > ul.status.left > li.vote').text
    comment = post.select_one('#container > div.wrap.articles > article > a > ul.status.left > li.comment').text
    scrap = post.select_one('#container > div.wrap.articles > article > a > ul.status.left > li.scrap').text
    
    data_dict = {
        "제목":title,
        "내용":contents,
        "공감":vote,
        "댓글수":comment,
        "스크랩수": scrap
                 }
    return data_dict
```

```python
def main():
    browser = _browser()
    browser.get('https://everytime.kr/login')
    login_evertime(browser, '', '')
    page_source = _click_hot_post(browser, 'second')
    data_dict = data_parser(page_source)
    print(data_dict)
```

```python
main()
```
